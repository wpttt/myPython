```python
#!/bin/python3

import os
import pandas as pd
import numpy as np

os.chdir('/Users/wpt/python/Rm')
#eqcatalog = './Rm2019.txt'
eqcatalog = './eqt.eqt'
eqcat = pd.read_csv(eqcatalog, sep='\n', header=None, encoding='gbk')
# eqcat.columns = ['eq']
# if 'ML' in eqcat['eq'][0] or 'MS' in eqcat['eq'][0]:
#     print('yes')
# else:
#     print('no')
eqcat
```

```python
eqcat['year'] = eqcat.apply(lambda x: x[0][0:4], axis=1)
eqcat['month'] = eqcat.apply(lambda x: x[0][4:6], axis=1)
eqcat['day'] = eqcat.apply(lambda x: x[0][6:8], axis=1)
eqcat['hour'] = eqcat.apply(lambda x: x[0][8:10], axis=1)
eqcat['minute'] = eqcat.apply(lambda x: x[0][10:12], axis=1)
eqcat['second'] = eqcat.apply(lambda x: x[0][12:16], axis=1)
eqcat['latitude'] = eqcat.apply(lambda x: x[0][16] + str(float(x[0][17:19]) + round(float(x[0][19:21])/60,2)), axis=1)
eqcat['longitude'] = eqcat.apply(lambda x: x[0][21] + str(float(x[0][22:25]) + round(float(x[0][25:27])/60,2)), axis=1)
eqcat['magnitude'] = eqcat.apply(lambda x: x[0][30:33],axis=1)
eqcat['depth'] = eqcat.apply(lambda x: x[0][33:37],axis=1)
eqcat['station'] = eqcat.apply(lambda x: x[0][37:42],axis=1)
eqcat['location'] = eqcat.apply(lambda x: x[0][42:],axis=1)
eqcat
```
